3. Utilizando el archivo figuras/dog_backpack.png realizar la rotación cada 60 grados utilizando imutils.rotate e imutils.rotate_bound

```python
img = cv.imread('figuras/dog_backpack.png')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
img_rotated = imutils.rotate(img, angle=60) # Rotar 60 grados a la derecha
img_rotated_bound = imutils.rotate_bound(img, angle= -60) # Rotar 60 grados a la derecha y sin perder bordes de la imagen

plt.figure(figsize=(8,6))
plt.subplot(1,2,1)
plt.title('Imagen rotate 60 grados')
plt.imshow(img_rotated)

plt.subplot(1,2,2)
plt.title('Imagen rotate_bound 60 grados')
plt.imshow(img_rotated_bound)
plt.show()  
```

4. Utilizando el archivo figuras/monedas.jpeg, obtenga los contornos como se muestra a continuacion
    
</div> 

<img align="left" width="300"  float= "none" align="middle" src="figuras/mx_coins.png">

```python
img = cv.imread('figuras/monedas.jpeg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
img_gris = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img_gauss = cv.GaussianBlur(img_gris, (3, 3), 0)
img_thr = cv.threshold(img_gauss, 170, 255, cv.THRESH_BINARY_INV)[1]
contornos, _ = cv.findContours(img_thr, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
cv.drawContours(img, contornos, -1, (255, 0, 0), 3)
plt.imshow(img)
plt.show()
```
