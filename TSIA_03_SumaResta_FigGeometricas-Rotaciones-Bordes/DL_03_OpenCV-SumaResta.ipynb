# 0. Librerías básicas

```python
#Librerías básicas de esta notebook
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv
```

# Parte II

### 1.Operaciones aritméticas

Las imágenes son arreglos de NumPy almacenados como enteros de 8 bits sin signo (unidad8) con valores en el rango [0, 255]; al usar las funciones de add/subtract en OpenCV, estos valores se *acoplarán* a este rango, incluso si caen fuera del rango [0, 255] después de aplicar la operación

```python
#Funciones de OpenCV
#add y subtract
added = cv.add(np.uint8([200]), np.uint8([100])) #sumar
subtracted = cv.subtract(np.uint8([50]), np.uint8([100])) #restar
print("Suma: {}".format(added))
print("Resta : {}".format(subtracted))
```

```python
#Cargar la imagen figuras/lago.jpg
img = cv.imread('figuras/lago.jpg')
#COnvertir a RGB
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
#Mostrar la imagen
plt.imshow(img)
plt.show()
img[0]
```

### 2. Aumentar las intensidad

```python
# Aumentar las intensidades de píxeles en nuestra imagen en 100
M = np.ones(img.shape, dtype = "uint8") * 100
img_suma = cv.add(img, M)
plt.imshow(img_suma)
plt.show()
img_suma[0]
```

### 3. Disminuir la intensidad

```python
# Disminuir la intensidade de píxeles en nuestra imagen con 50
M = np.ones(img.shape, dtype = "uint8") * 100
img_resta = cv.subtract(img, M)
plt.imshow(img_resta)
plt.show()
img_resta[0]
```

### 4. Alta y baja intensidad

```python
plt.figure(figsize=(5,10))
plt.subplot(3,1,1)
plt.title("Imagen Original")
plt.imshow(img)
plt.subplot(3,1,2)
plt.title("Imagen Suma")
plt.imshow(img_suma)
plt.subplot(3,1,3)
plt.title("Imagen Resta")
plt.imshow(img_resta)
plt.show()
```
