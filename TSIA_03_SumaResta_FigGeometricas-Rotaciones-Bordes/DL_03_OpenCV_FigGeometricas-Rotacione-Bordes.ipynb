# 0. Librerías básicas

```python
#Librerías básicas de esta notebook
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv
```

# 1. Figuras geométricas


### Creación de una imagen en color negro

```python
# Imagen color negro de 512 x 512 
img_base = np.zeros(shape=(512, 512, 3), dtype=np.uint8)
plt.imshow(img_base)
plt.show()
```

#### Rectangulo

```python
#cv.rectangle(imagen,origen,final,color,grosor de linea)
# Imagen color negro de 512 x 512 
img_base = np.zeros(shape=(512, 512, 3), dtype=np.uint8)
cv.rectangle(img_base,(100,100),(400,200),(0,255,0), thickness=6) #x es 100, y es 100,
cv.rectangle(img_base,(100,200),(200,400),(255,0,0), thickness=6) #x es 100, y es 200.
plt.imshow(img_base)
plt.show()
```

#### Circulo

```python
#cv.circle(imagen, centro, radio, color, grosor de linea)
# Imagen color negro de 512 x 512 
img_base = np.zeros(shape=(512, 512, 3), dtype=np.uint8)
cv.circle(img_base, (256, 256), 100, (0, 0, 255), thickness=6) # centro en (256, 256), radio 100
cv.circle(img_base, (256, 256), 50, (255, 255, 255), thickness=-1) # centro en (256, 256), radio 100
plt.imshow(img_base)
plt.show()
```

#### Linea

```python
#cv.line(imagen, inicio, fin, color, grosor de linea)
img_base = np.zeros(shape=(512, 512, 3), dtype=np.uint8)
cv.line(img_base, (0, 0), (512, 512), (0, 255, 0), thickness=6) # desde (0, 0) hasta (500, 500)
cv.line(img_base, (512, 0), (0, 512), (255, 0, 0), thickness=6) # desde (512, 0) hasta (0, 512)
plt.imshow(img_base)
plt.show()
```

#### Texto

```python
font = cv.FONT_HERSHEY_SIMPLEX
lineType=cv.LINE_AA
#cv.putText(imagen,texto,origen, fuente,tamaño de la fuente,color,grosor, tipo de linea)

img_base = np.zeros(shape=(512, 512, 3), dtype=np.uint8)
cv.putText(img_base, "Hola Mundo", (100, 400), fontFace=font, fontScale=2, color=(0, 255, 0), thickness=3, lineType=lineType)
cv.putText(img_base, "Python", (100, 300), fontFace=font, fontScale=3.5, color=(0, 0, 255), thickness=3, lineType=lineType)
plt.imshow(img_base)
plt.show()
```


#### Polígono 

```python
#cv.polylines(imagen,puntos,Cerrada o abierta,color y grosor)
img_base = np.zeros(shape=(512, 512, 3), dtype=np.uint8)
puntos = np.array([[100, 200], [400, 200], [400, 300], [200, 400]], np.int32)
cv.polylines(img_base, [puntos], isClosed=True, color=(255, 255, 0), thickness=5)
plt.imshow(img_base)
plt.show()
```

## 4. Detección de contornos

```python
#Visualizar figuras/coins.png
img = cv.imread('figuras/coins.png')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
#Pasar a escala de grises
img = cv.imread('figuras/coins.png')
img_gris = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(img_gris, cmap='gray')
plt.show()
```

```python
#Aplicar el filtro GaussianBlur
img = cv.imread('figuras/coins.png')
img_gris = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img_gauss = cv.GaussianBlur(img_gris, (3,3), 0)
plt.imshow(img_gauss, cmap='gray')
plt.show()
```

```python
#Aplicar la binarización con la opción cv.THRESH_BINARY_INV
img = cv.imread('figuras/coins.png')
img_gris = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img_gauss = cv.GaussianBlur(img_gris, (3,3), 0)
img_thr = cv.threshold(img_gauss, 150, 255, cv.THRESH_BINARY_INV)[1] #retorna una tupla, el segundo elemento es la imagen binarizada, aparte invierte blanco y negro _INV
plt.imshow(img_thr, cmap='gray')
plt.show()
```

```python
#Encontrar los contornos con cv.findContours(IMAGEN-BN, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE )
contornos, _ = cv.findContours(img_thr, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
len(contornos)
```

```python
# Visualizar los contornos  con drawContours
img = cv.imread('figuras/coins.png')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
cv.drawContours(img, contornos, -1, (0, 255, 0), 2) #-1 es para dibujar todos los contornos, 2 es el grosor
plt.imshow(img)
plt.show()
```

```python
# Primer contorno
img = cv.imread('figuras/coins.png')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
cv.drawContours(img, contornos, 0, (255, 0, 0), 2) #0 es para el primer contorno, 2 es el grosor
plt.imshow(img)
plt.show()
```
